```python
import pandas as pd
import os
from pathlib import Path
```

```python
pwd = Path('/Users/a19336136/rnc_cluster/sem_cluster_last/')
```

```python
df_clusters = pd.DataFrame(columns=['identifier', 'cluster']) 
```

```python
for file in os.listdir('/Users/a19336136/rnc_cluster/sem_cluster_last/clusters'):
    df = pd.read_csv(os.path.join(pwd, f'/Users/a19336136/rnc_cluster/sem_cluster_last/clusters/{file}'), sep='\t')
    df_clusters = pd.concat([df_clusters, df])
```

```python
df_clusters.reset_index(inplace=True)
df_clusters.drop(columns=['index'], inplace=True)
```

```python
df_uses = pd.DataFrame()
```

```python
for file in os.listdir(os.path.join(pwd, 'data')):
    try:
        df = pd.read_csv(os.path.join(pwd, f'data/{file}/uses.csv'), sep='\t')
        df_uses = pd.concat([df_uses, df])
    except:
        pass
```

```python
df_uses
```

```python
df_russe = pd.merge(df_uses[['identifier', 'context', 'indexes_target_token']], df_clusters, right_on='identifier', left_on='identifier')
```

```python
df_russe.drop_duplicates(inplace=True)
```

```python
df_russe['identifier'] = df_russe['identifier'].apply(lambda x: x.split('_')[1])
```

```python
df_russe['indexes_target_token'] = df_russe['indexes_target_token'].apply(lambda x: x.replace(':', '-'))
```

```python
df_russe.rename(columns={"identifier": "word", "cluster": "gold_sense_id", "indexes_target_token": "positions"}, inplace=True)
```

```python
df_russe.reset_index(inplace=True)
df_russe.drop(columns=['index'], inplace=True)
```

```python
df_russe['predict_sense_id'] = [''] * df_russe.shape[0]
```

```python
df_russe.groupby('word')['gold_sense_id'].nunique()
```

```python
df_russe.to_csv("sem_cluster_data.tsv", sep='\t')
```
